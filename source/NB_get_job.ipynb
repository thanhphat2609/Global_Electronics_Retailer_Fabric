```python
task_name_param = ""
```

```python
import json
from pyspark.sql.functions import col

# Define file_path_warehouse
file_path_warehouse = "abfss://Global_Electronics_Retailer_Meta@onelake.dfs.fabric.microsoft.com/WH_Metadata.Datawarehouse/Tables/metadata/config_table"

df = spark.read.format("delta").load(file_path_warehouse)

df_task_run = df.filter(col("task_name") == f"{task_name_param}").orderBy("Task_ID", ascending = True)

# display(df)
```

```python
# Get source_to_bronze job
json_array = df_task_run.toJSON().collect()

# display df
print(json_array)
```

```python
# Initialize an empty list to store the JSON objects
json_objs = []

# Convert each JSON string to a JSON object
for json_str in json_array:
    json_obj = json.loads(json_str)
    json_objs.append(json_obj)

# Get table_loaded from result
table_loaded = []

# Get data for source_to_bronze
if task_name_param == "source_to_bronze":
    for json_obj in json_objs:
        result = { \
            "task_id": json_obj["task_id"], \
            "task_name": json_obj["task_name"], \
            "source_connection": json_obj["source_connection"], \
            "source_database": json_obj["source_database"], \
            "source_schema": json_obj["source_schema"], \
            "source_table": json_obj["source_table"], \
            "target_folder": json_obj["target_folder"], \
            "load_type": json_obj["load_type"], \
            "phase": json_obj["phase"] \
        }
        table_loaded.append(result)

# Get data for bronze_to_silver
elif task_name_param == "bronze_to_silver":
    for json_obj in json_objs:
        result = { \
            "task_id": json_obj["task_id"], \
            "task_name": json_obj["task_name"], \
            "source_folder": json_obj["source_folder"], \
            "target_schema": json_obj["target_schema"], \
            "target_table": json_obj["target_table"], \
            "phase": json_obj["phase"]
        }
        table_loaded.append(result)

# Print table_loaded
print(table_loaded)
```
