```python
# Import python file as module
sc.addPyFile("abfss://Global_Electronics_Retailer_Meta@onelake.dfs.fabric.microsoft.com/LH_Python_Files.Lakehouse/Files/FabricUDFManagement.py")
from FabricUDFManagement import *

# Instanct of FabricUtils
fabricUtils = FabricUtils()

# Import more modules
from pyspark.sql.functions import col
import pandas as pd
import sempy.fabric as fabric
import json

# Instantiate the client
client = fabric.FabricRestClient()
```

```python
workspaceinfo_df = fabricUtils.get_workspace_info(fabric, spark)

workspaceinfo_df = workspaceinfo_df.withColumnRenamed("Id", "workspaceid")

# display(workspaceinfo_df)
```

```python
workspaceitem_df = fabricUtils.get_items_in_workspace(fabric, client, pd, spark)

# display(workspaceitem_df)
```

```python
# Get full information about workspace and item of workspace
df_full_info = workspaceinfo_df.join(workspaceitem_df, on = ["workspaceid"], how = "inner") \
                               .select(["workspaceid", "Name", "id", "displayName", "type"])

df_full_info = df_full_info.withColumnRenamed("workspaceid", "WorkspaceId") \
                           .withColumnRenamed("Name", "WorkspaceName") \
                           .withColumnRenamed("id", "ElementId") \
                           .withColumnRenamed("displayName", "ElementName") \
                           .withColumnRenamed("type", "ElementType")

# display(df_full_info)
```

```python
# Get lakehouse for Source or Destination
lakehouse = df_full_info.filter(col("ElementType") == "Lakehouse").select("WorkspaceId", "ElementId", "ElementName")

# Convert to json
lakehouse_json_array = lakehouse.toJSON().collect()

# Convert to JsonObj
for json_str in lakehouse_json_array:
    json_obj = json.loads(json_str)

# Get values from each JSON object
result = { \
    "WorkspaceId": json_obj["WorkspaceId"], \
    "ElementId": json_obj["ElementId"], \
    "ElementName": json_obj["ElementName"]
}

# print(result)
```
