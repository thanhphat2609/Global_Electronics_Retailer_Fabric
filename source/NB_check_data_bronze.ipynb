```python
# Receive parameters
executionYear = ""
executionMonth = ""
executionDay = ""
pipelineruntime = ""
bronze_action = ""
workspaceid = ""
lakehouseid = ""
```

```python
import json

# Import pyfile as module
sc.addPyFile("abfss://Global_Electronics_Retailer_Meta@onelake.dfs.fabric.microsoft.com/LH_Python_Files.Lakehouse/Files/FabricUDFManagement.py")
from FabricUDFManagement import *

# Create an instance
pipelineLogger = PipelineLogger()

from pyspark.sql.functions import *
import pyspark.sql.types as types

# Create base path for Lakehouse
base_lakehouse_path = f"abfss://{workspaceid}@onelake.dfs.fabric.microsoft.com/{lakehouseid}/"

# Concat executionDate
executionDate = f"{executionYear}-{executionMonth}-{executionDay}"
```

```python
# Import lib
from pyspark.sql.functions import col, concat
import traceback
import logging

# Status
error = ""
status = ""

# Inserted, Updated
numInserted = 0
numUpdated = 0

# Task
task_id = 0
task_name = ""
phase = ""
folder = ""


# Convert param from pipeline
bronze_action_json_array = json.loads(bronze_action)


for bronze_action_json in bronze_action_json_array:

    # Load in to json object
    bronze_check = json.loads(bronze_action_json)


    # Start time for check
    start_time = spark.sql(''' SELECT from_utc_timestamp(CURRENT_TIMESTAMP(), 'Asia/Ho_Chi_Minh') as current_time ''') \
                              .collect()[0]["current_time"].strftime('%Y-%m-%d %H:%M:%S')

    task_id = bronze_check["task_id"]
    task_name = bronze_check["task_name"]
    phase = bronze_check["phase"]
    folder = bronze_check["target_folder"]
    src_rows_read = int(bronze_check["src_rows_read"])
    numInserted = int(bronze_check["numInserted"])

    column_missing = ""
    column_null_str = ""

    # Check exits file in Files of Lakehouse
    try:

        # Define file_path
        file_path = f"{base_lakehouse_path}/Files/Bronze/{folder}/{executionYear}/{executionMonth}/{executionDay}/*.parquet"

        # Load to dataframe
        df_bronze = spark.read.format("parquet").load(file_path)

    except:
        error = traceback.format_exc()
        status = "Failed"
    
    else:
        error = ""
        status = "Success"


    end_time = spark.sql(''' SELECT from_utc_timestamp(CURRENT_TIMESTAMP(), 'Asia/Ho_Chi_Minh') as current_time ''') \
                                .collect()[0]["current_time"].strftime('%Y-%m-%d %H:%M:%S')

    # Call function
    df_log = pipelineLogger.log_data(pipelineruntime, task_id, task_name, \
                                      start_time, end_time, src_rows_read, numInserted, numUpdated, \
                                      column_missing, column_null_str, error, status, types, spark \
                                    )

    # Save as parquet files
    df_log.write.mode("append").format("parquet").save(f"{base_lakehouse_path}/Files/Log_SourcetoBronze/{executionDate}/")
```
